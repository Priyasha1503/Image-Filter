```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import convolve #for blur image conversion
def convert_rgb_to_grey(img_array):
    greyscale=np.dot(img_array[...,:3],[0.2989, 0.5870, 0.1140])
    return greyscale
def convert_rgb_to_negative(img_array):
    #Invert the image
    inverted_img=255-img_array
    plt.imshow(inverted_img)
    plt.title("Inverted(Negative) Image")
    plt.axis('off')
    plt.show()

def convert_to_blur(img_array):#here img _array is actually grey_array,c in func call
  #define a 3X3 average kernel
  blur_kurnel=np.ones((3,3))/9.0
  blurred_image= convolve(img_array,blur_kurnel)
  plt.imshow(blurred_image,cmap="gray")
  plt.title("Blurred Image")
  plt.axis('off')
  plt.show()

def convert_rgb_to_sobel(img_array):
  sobel_x=np.array([
      [-1,0,1],
      [-2,0,2],
      [-1,0,1]
  ])
  sobel_y=np.array([
      [-1,-2,-1],
      [0,0,0],
      [1,2,1]
  ])
  edge_x=convolve(img_array,sobel_x)
  edge_y=convolve(img_array,sobel_y)
  gradient_magnitude=np.sqrt(edge_x**2+edge_y**2)
  gradient_magnitude*=255.0/gradient_magnitude.max()
  plt.imshow(gradient_magnitude,cmap="gray")
  plt.title("Sobel Edge Detection (Pencil Sketch)")
  plt.axis("off")
  plt.show()


###IMAGE - ORIGINAL
img=Image.open("littlekrishna.jpg").convert("RGB")
img_array=np.array(img)
plt.imshow(img_array)
plt.title("Original picture")
plt.axis('off')
plt.show()
###GREY IMAGE - converting and plotting the image...
grey_array=convert_rgb_to_grey(img_array)
plt.imshow(grey_array,cmap="gray")
plt.title("GREY SCALE IMAGE")
plt.axis('off')
plt.show()
###INVERTED IMAGE OR NEGATIVE IMAGE
convert_rgb_to_negative(img_array)
###BLUR IMAGE
convert_to_blur(grey_array) #rgb cannot be directly converted to blur,u have to convert
#rgb to blur first, to convert it to blur,since we have already converted it before..using that grey_array now
### SOBEL EDGE DETECTION (PENCIL SKETCH)
convert_rgb_to_sobel(grey_array)

```
